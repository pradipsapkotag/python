### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

```python
data_url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
df = pd.read_csv(data_url,sep='\t')
df
```

### Step 3. Assign it to a variable called chipo.

```python
chipo = df
chipo
```

### Step 4. How many products cost more than $10.00?

```python
df['item_price'] = df['item_price'].str.replace('$', '').astype(float)
df['item_price']
```

```python
df.item_price>10.00
```

```python
(df.item_price>10.00).sum()
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
df_item_price = df.groupby(by ="item_name")[["item_price"]].min()
df_item_price.sort_values(by='item_price',ascending= False)
```

### Step 6. Sort by the name of the item

```python
df = df.sort_values(by = "item_name",ascending=True).reset_index()
df
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
df_item_price.sort_values(by='item_price',ascending= False)
```

```python
df_item_price.item_price.max()
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
df.loc[df['item_name']=='Veggie Salad Bowl'].quantity.sum()
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
canned_soda_df =  df.loc[df['item_name']=='Canned Soda']
canned_soda_df_more_quantity = canned_soda_df.loc[canned_soda_df['quantity']>1]
canned_soda_df_more_quantity.shape[0]
```
